## Read Input

```python
with open("./inputs/day09.txt") as f:
    input_string = f.read()
```

## Parse Input

```python
def parse(text: str) -> list[tuple[int, int]]:
    lines = text.strip().splitlines()
    red_tiles = []
    for line in lines:
      split = line.split(",")
      red_tiles.append((int(split[0]), int(split[1])))
    return red_tiles

parsed = parse(input_string)
```

## Solve

```python
def solve(data: list[tuple[int, int]]) -> int:
    pairs = [(a, b) for i, a in enumerate(data) for b in data[i+1:]]
    larges_area = 0
    for a, b in pairs:
        width = abs(a[0] - b[0]) + 1
        height = abs(a[1] - b[1]) + 1
        area = width * height
        if area > larges_area:
            larges_area = area
    return larges_area

result = solve(parsed)
print(result)
```

```python
def solve_part2(data: list[tuple[int, int]]) -> int:
    pairs = [(a, b) for i, a in enumerate(data) for b in data[i+1:]]
    largest_area = 0
    largest_a = None
    largest_b = None
    for a, b in pairs:
        width = abs(a[0] - b[0]) + 1
        height = abs(a[1] - b[1]) + 1
        area = width * height

        if area > largest_area and is_only_red_and_green_tiles(a, b, data):
            largest_area = area
            largest_a = a
            largest_b = b
    
    return largest_area, largest_a, largest_b

def is_only_red_and_green_tiles(a: tuple[int, int], b: tuple[int, int], data: list[tuple[int, int]]) -> bool:
    # since both the rectangle and the edges are axis-aligned, we can do a simple axis-aligned bounding box check
    # if any edge is within the rectangle (not on the boundary), return False
    
    rect_min_x = min(a[0], b[0])
    rect_max_x = max(a[0], b[0])
    rect_min_y = min(a[1], b[1])
    rect_max_y = max(a[1], b[1])

    n = len(data)
    for i in range(n):
        p1 = data[i]
        p2 = data[(i+1)%n]
        edge_min_x = min(p1[0], p2[0])
        edge_max_x = max(p1[0], p2[0])
        edge_min_y = min(p1[1], p2[1])
        edge_max_y = max(p1[1], p2[1])
        if (rect_min_y < edge_max_y and edge_min_y < rect_max_y
            and rect_min_x < edge_max_x and edge_min_x < rect_max_x):
            return False
    return True

result, a, b = solve_part2(parsed)
print(result)
```
