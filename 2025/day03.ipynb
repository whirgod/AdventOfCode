<p>The batteries are arranged into <em>banks</em>; each line of digits in your input corresponds to a single bank of batteries. Within each bank, you need to turn on <em>exactly two</em> batteries; the joltage that the bank produces is equal to the number formed by the digits on the batteries you've turned on. For example, if you have a bank like <code>12345</code> and you turn on batteries <code>2</code> and <code>4</code>, the bank would produce <code>24</code> jolts. <p>You'll need to find the largest possible joltage each bank can produce. In the above example:</p>
<ul>
<li>In <code><em>98</em>7654321111111</code>, you can make the largest joltage possible, <em><code>98</code></em>, by turning on the first two batteries.</li>

<li>In <code>2342342342342<em>78</em></code>, you can make <em><code>78</code></em> by turning on the last two batteries (marked <code>7</code> and <code>8</code>).</li>
<li>In <code>818181<em>9</em>1111<em>2</em>111</code>, the largest joltage you can produce is <em><code>92</code></em>.</li>
</ul>
<p>The total output joltage is the sum of the maximum joltage from each bank, so in this example, the total output joltage is <code>98</code> + <code>89</code> + <code>78</code> + <code>92</code> = <code><em>357</em></code>.</p>
<p>There are many batteries in front of you. Find the maximum joltage possible from each bank; <em>what is the total output joltage?</em></p>

## Read Input

```python
with open("./inputs/day03.txt") as f:
    input_string = f.read()
```

## Parse Input

```python
def parse(text: str) -> list[str]:
    return text.strip().splitlines()

parsed = parse(input_string)
```

## Solve Part 1

```python
def solve_part1(data: list[str]) -> int:
    return sum(get_max_joltage(line) for line in data)

def get_max_joltage(line: str) -> int:
    if (len(line) < 2):
        raise ValueError("Line must have at least two digits")
    
    first_digit = line[0]
    first_digit_index = 0
    for (index, digit) in enumerate(line[1:-1]):
        if digit > first_digit:
            first_digit = digit
            first_digit_index = index + 1
            if digit == '9':
                break

    second_digit = max(line[first_digit_index + 1:])
    return int(first_digit + second_digit)
    



result = solve_part1(parsed)
print(result)
```

<li>In the digit sequence <code><em>81111111111</em>111<em>9</em></code>, the largest joltage can be found by turning on everything except some <code>1</code>s, producing <code><em>811111111119</em></code>.</li>

<p>The total output joltage is now much larger: <code>987654321111</code> + <code>811111111119</code> + <code>434234234278</code> + <code>888911112111</code> = <code><em>3121910778619</em></code>.</p>

## Solve Part 2

```python
def solve_part2(data: list[str]) -> int:
    return sum(get_max_joltage_general(line, 12) for line in data)

def get_max_joltage_general(line: str, number_of_digits: int) -> int:
    if (len(line) < number_of_digits):
        raise ValueError("Line must have at least two digits")
    
    digits = []
    indices = []

    for i in range(number_of_digits):
        if i == 0:
            digits.append(line[0])
            indices.append(0)
        else:
            digits.append(line[indices[-1] + 1])
            indices.append(indices[-1] + 1)
        
        for index in range(indices[-1] + 1, len(line) - (number_of_digits - i - 1)):
            digit = line[index]
            if digit > digits[-1]:
                digits[-1] = digit
                indices[-1] = index
                if digit == '9':
                    break

    return int(''.join(digits))
    



result = solve_part2(parsed)
print(result)
```
