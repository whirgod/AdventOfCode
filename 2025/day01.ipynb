## Read Input

```python
with open("./inputs/day01.txt") as f:
    input_string = f.read()
```

## Parse Input

```python
from typing import Any

def parse(text: str) -> Any:
    lines = text.strip().splitlines()
    return [parse_line(line) for line in lines if line.strip()]

def parse_line(line: str) -> Any:
    if line[0] == 'L':
        sign = -1
    elif line[0] == 'R':
        sign = 1
    else:
        raise ValueError(f"Invalid line format: {line}")
    
    value = int(line[1:])
    return sign * value


parsed = parse(input_string)
```

## Solve

```python
def solve(data: Any) -> int:
    value = 50
    counter = 0
    for change in data:
        value += change
        value %= 100
        if value == 0:
            counter += 1
    
    return counter

result = solve(parsed)
print(result)
```
