## Read Input

```python
with open("./inputs/day08.txt") as f:
    input_string = f.read()
```

## Parse Input

```python
def parse(text: str) -> list[tuple[int, int, int]]:
    splits = [line.split(',') for line in text.strip().splitlines()]
    result = []
    for split in splits:
        result.append((int(split[0]), int(split[1]), int(split[2])))
    return result

parsed = parse(input_string)
```

## Solve

```python
from math import sqrt


class JunctionBox:
    def __init__(self, x: int, y: int, z: int, name: str | None = None):
        self.x = x
        self.y = y
        self.z = z
        self.circuit = Circuit({self})
        if name is None:
            self.name = f"JunctionBox({x}, {y}, {z})"
        else:
            self.name = name
    
    def set_circuit(self, circuit: Circuit):
        self.circuit = circuit
    
    def __str__(self):
        return self.name
    
    def __repr__(self):
        return str(self)

class Circuit:
    def __init__(self, junction_boxes: set['JunctionBox']):
        self.junction_boxes = junction_boxes
    
    def merge(self, other: 'Circuit') -> 'Circuit':
        merged_boxes = self.junction_boxes.union(other.junction_boxes)
        merged_circuit = Circuit(merged_boxes)
        for box in merged_boxes:
            box.set_circuit(merged_circuit)
        return merged_circuit
    
    def __str__(self):
        return str(self.junction_boxes)
    
    def __repr__(self):
        return str(self)
    
    def __len__(self):
        return len(self.junction_boxes)

def print_circuits(data: list['JunctionBox']):
    circuits = {box.circuit for box in data}
    print(circuits)
    print()

def solve(data: list['JunctionBox'], number_of_connections: int) -> int:
    pairs = [(a, b) for i, a in enumerate(data) for b in data[i+1:]]
    pairs.sort(key=lambda pair: sqrt((pair[0].x - pair[1].x)**2 + (pair[0].y - pair[1].y)**2 + (pair[0].z - pair[1].z)**2))
    
    # print_circuits(data)

    for a, b in pairs[:number_of_connections]:
        a.circuit.merge(b.circuit)
        # print_circuits(data)

    circuits = list({box.circuit for box in data})
    circuits.sort(key=lambda c: len(c.junction_boxes), reverse=True)
    return len(circuits[0]) * len(circuits[1]) * len(circuits[2])

junction_boxes = [JunctionBox(x, y, z, f'JB{i}') for i, (x, y, z) in enumerate(parsed)]
result = solve(junction_boxes, 1000)
print(result)
```
